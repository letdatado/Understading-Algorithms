# POINT

```python
class Point:
    def __init__(self, x: int, y: int):
        self.x = x
        self.y = y

    def get_x(self) -> int:
        return self.x

    def get_y(self) -> int:
        return self.y

    def get_distance(self, target: "Point") -> float:
        diff_x = target.get_x() - self.x
        diff_y = target.get_y() - self.y
        return (diff_x ** 2 + diff_y ** 2) ** 0.5

    def __str__(self) -> str:
        return f"Point({self.x},{self.y})"
```

# GRAPH

```python
class Graph:
    def __init__(self, vertices, edges):
        self.vertices = vertices
        self.edges = edges
    def __str__(self) -> str:
        return f"Point({self.vertices},{self.edges})"
```

```python
# Example Vertices
point_alpha = Point(1, 2)
point_beta = Point(4, 3)
point_gamma = Point(7, 8)

vertices = [point_alpha, point_beta, point_gamma]

# Example Edges
edges = [
    (0, 1, point_alpha.get_distance(point_beta)),
    (1, 2, point_beta.get_distance(point_gamma)),
    (0, 2, point_alpha.get_distance(point_gamma))
]
```

# KRUSKAL MST

```python
from typing import List, Tuple
```

```python
graph = Graph(list_of_all_points, list_of_edges)
graph
```

```python
parent = list(range(len(graph.vertices)))
print(parent)
```

```python
def find_root(vertex_index: int) -> int:
    if parent[vertex_index] != vertex_index:
        parent[vertex_index] = find_root(parent[vertex_index])
    return parent[vertex_index]

def connect_points(vertex_a_index: int, vertex_b_index: int) -> bool:
        root_a = find_root(vertex_a_index)
        root_b = find_root(vertex_b_index)

        if root_a != root_b:
            parent[root_b] = root_a
            return True
        else:
            return False

mst_edges: List[Tuple[Point, Point, float]] = []
```

```python
def get_weight(edge: Tuple[int, int, float]) -> float:
    weight_index = 2
    return edge[weight_index]

sorted_edges = sorted(graph.edges, key=get_weight)
```

```python
# Peak
sorted_edges[:5]
```

```python
for vertex_a_index, vertex_b_index, weight in sorted_edges:
    if connect_points(vertex_a_index, vertex_b_index):
        start_point = graph.vertices[vertex_a_index]
        end_point = graph.vertices[vertex_b_index]
        mst_edges.append((start_point, end_point, weight))

    if len(mst_edges) == len(graph.vertices) - 1:
        break
```
